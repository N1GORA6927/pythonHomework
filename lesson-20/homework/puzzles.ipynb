```python
import pandas as pd
```

```python
# Data for df1
data1 = {'ProductID': [1, 2, 3],
         'ProductName': ['Widget', 'Gadget', 'Doodad'],
         'Category': ['Tools', 'Electronics', 'Toys']}

df1 = pd.DataFrame(data1)

# Data for df2
data2 = {'ProductID': [1, 2, 4],
         'Price': [19.99, 29.99, 15.99],
         'Stock': [100, 200, 150]}

df2 = pd.DataFrame(data2)

```

```python
df1
```

```python
#1
df1.merge(df2,on='ProductID',how='outer')
```

```python
#2
# Data for customers
data_customers = {'CustomerID': [1, 2, 3],
                  'CustomerName': ['Alice', 'Bob', 'Charlie'],
                  'Country': ['USA', 'UK', 'Canada']}

customers = pd.DataFrame(data_customers)

# Data for orders
data_orders = {'OrderID': [101, 102, 103],
               'CustomerID': [1, 1, 3],
               'OrderDate': ['2025-01-15', '2025-02-15', '2025-03-15']}

orders = pd.DataFrame(data_orders)
```

```python
customers.merge(orders,on='CustomerID',how='left')
```

```python
#3# Data for sales_jan
data_jan = {'Date': ['2025-01-01', '2025-01-02', '2025-01-03'],
            'ProductID': [1, 2, 3],
            'Sales': [10, 20, 30]}

sales_jan = pd.DataFrame(data_jan)

# Data for sales_feb
data_feb = {'Date': ['2025-02-01', '2025-02-02', '2025-02-03'],
            'ProductID': [1, 2, 3],
            'Sales': [15, 25, 35]}

sales_feb = pd.DataFrame(data_feb)

# Data for sales_mar
data_mar = {'Date': ['2025-03-01', '2025-03-02', '2025-03-03'],
            'ProductID': [1, 2, 3],
            'Sales': [20, 30, 40]}

sales_mar = pd.DataFrame(data_mar)
```

```python
res=pd.concat([sales_jan,sales_feb,sales_mar])
res.sort_values('Date')
```

```python
#4
# Data for df1
data_emp1 = {'EmployeeID': [1, 2, 3],
             'Name': ['John', 'Jane', 'Doe'],
             'Department': ['HR', 'Finance', 'IT']}

df1 = pd.DataFrame(data_emp1)

# Data for df2
data_emp2 = {'EmployeeID': [1, 3, 4],
             'Salary': [50000, 60000, 55000],
             'JoinDate': ['2023-01-10', '2024-02-15', '2025-03-01']}

df2 = pd.DataFrame(data_emp2)
```

```python
df1.merge(df2,on='EmployeeID',how='left')
```

```python
# Data for students
data_students = {'StudentID': [1, 2, 3],
                 'Name': ['Anna', 'Bill', 'Cathy'],
                 'Major': ['Math', 'Physics', 'Chemistry']}

students = pd.DataFrame(data_students)

# Data for courses
data_courses = {'CourseID': [101, 102, 103],
                'CourseName': ['Calculus', 'Mechanics', 'Organic Chemistry'],
                'Credits': [4, 3, 4]}

courses = pd.DataFrame(data_courses)

# Data for enrollments
data_enrollments = {'StudentID': [1, 1, 2],
                    'CourseID': [101, 102, 103],
                    'Grade': ['A', 'B+', 'A-']}

enrollments = pd.DataFrame(data_enrollments)
```

```python
result=students.merge(enrollments,on='StudentID',how='inner')
final=result.merge(courses,on='CourseID',how='left')
final
```
